```python
import pandas as pd
import numpy as np
```

```python
file2=pd.read_csv("D:\IPL_WinPrediction\IPL_Ball_by_Ball_2008_2022.csv")
file2
```

```python
matches_played = file2.groupby('batter')['ID'].nunique().reset_index(name='MatchesPlayed')
matches_played.sort_values(by='batter',ascending=True)
matches_played
```

```python
run_total = file2.groupby('batter').apply(lambda x:np.sum(x['batsman_run'])).reset_index(name='Runs')
runs_total = run_total.sort_values(by='batter',ascending=True)
runs_total
```

```python
batting_avg = pd.merge(matches_played,runs_total,on='batter',how='inner')
batting_avg['BattingAverage'] = batting_avg['Runs']/batting_avg['MatchesPlayed']
batting_avg.sort_values(by='batter',ascending=True,inplace=True)
batting_avg
```

```python
total = file2.groupby('batter').apply(lambda x:np.sum(x['batsman_run'])).reset_index(name='Runs')
balls_faced=file2.groupby('batter').count()
balls_total = balls_faced['ballnumber'].reset_index(name='Balls')
batsman_runs_balls=pd.merge(total,balls_total,left_on='batter',right_on='batter',how='outer')
strike_rates = batsman_runs_balls.groupby(['Runs','batter']).apply(lambda x:((x['Runs'])/(x['Balls']))*100).reset_index(name='Strike Rate')
strike_rate=strike_rates.sort_values(by='batter',ascending=True)
strike_rate
```

```python
strike_rate.drop(['Runs'],axis=1,inplace=True)
```

```python
batting_perfo = pd.merge(batting_avg,strike_rate,on='batter',how='inner')
batting_perfo
```

```python
batting_perfo.drop(['MatchesPlayed','Runs','level_2'],axis=1,inplace=True)
batting_perfo
```

```python
runs=file2.groupby('bowler').apply(lambda x:np.sum(x['total_run'])).reset_index(name='Runs Conceded')
runs
```

```python
wicket_total = file2[file2['kind']!='run out']
wicket_totals = wicket_total.groupby('bowler').apply(lambda x:x['kind'].dropna()).reset_index(name='Wickets')
wickets_tot = wicket_totals.groupby('bowler').count().reset_index()
wickets_total = wickets_tot.sort_values(by='bowler') 
wickets_total
```

```python
bowling_avg = pd.merge(runs,wickets_total,on='bowler',how='outer')
bowling_avg.fillna(value=0,inplace=True)
bowling_avg['BowlingAverage'] = bowling_avg['Runs Conceded']/bowling_avg['Wickets']
bowling_avg['BowlingAverage'] = bowling_avg['BowlingAverage'].replace(np.inf,0)
bowling_avg
```

```python
runs=file2.groupby('bowler').apply(lambda x:np.sum(x['total_run'])).reset_index(name='Runs Conceded')
balls=file2.groupby('bowler').count()
bowled_balls=balls['ballnumber'].reset_index(name='Balls Bowled')
bowler_stats=pd.merge(runs,bowled_balls,left_on='bowler',right_on='bowler',how='outer')
economy_rates=bowler_stats.groupby(['bowler','Balls Bowled']).apply(lambda x:(((x['Runs Conceded'])/(x['Balls Bowled']))*6)).reset_index(name='Economy Rate')
economy_rate=economy_rates.sort_values(by='bowler',ascending=True)
economy_rate
```

```python
bowling_perfo = pd.merge(bowling_avg,economy_rate,on='bowler',how='inner')
bowling_perfo.fillna(value=0)
```

```python
bowling_perfo.drop(['level_1','Balls Bowled','level_2'],axis=1,inplace=True)
bowling_perfo
```

```python
player_perfo = pd.merge(batting_perfo,bowling_perfo,left_on='batter',right_on='bowler',how='outer')
player_perfo
```

```python
player_perfo['batter'].fillna(player_perfo['bowler'], inplace=True)
```

```python
player_perfo.fillna(value=0,inplace=True)
```

```python
player_perfo.drop(['bowler'],axis=1,inplace=True)
```

```python
player_perfo.rename({'batter':'Player'},axis=1,inplace=True)
player_perfo
```

```python
player_perfo.to_csv("D:\IPL_WinPrediction\IPL_PlayerPerformance.csv",index=False)
```

```python
def classify_players(row):
    if row['BattingAverage'] > 20 and row['BowlingAverage'] <30:
            return 'Allrounder'
    elif row['BattingAverage'] > 20:
        return 'Batsman'
    elif row['BowlingAverage'] <30:
        return 'Bowler'
    else:
        return 'Player'
```

```python
player_perfo['PlayerType'] = player_perfo.apply(classify_players, axis=1)
player_perfo
```

```python
player_type=player_perfo[['Player','PlayerType']].copy()
player_type
```
