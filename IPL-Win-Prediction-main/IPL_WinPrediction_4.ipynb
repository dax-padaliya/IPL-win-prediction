```python
def result(row):
    return 1 if row['Team1'] == row['WinningTeam'] else 0
```

```python
def T1HomeGround(row):
     return "Yes" if row['City'] == row['HomeGround_x'] else "No"
```

```python
def T2HomeGround(row):
     return "Yes" if row['City'] == row['HomeGround_y'] else "No"
```

```python
def battingfirst(row):
        if row['Team1'] == row['TossWinner']:
            if row['TossDecision'] == 'bat':
                return 1
            else:
                return 0
        elif row['Team2'] == row['TossWinner']:
            if row['TossDecision'] == 'bat':
                return 0
            else:
                return 1
```

```python
IPL_Final_Data['Result'] = IPL_Final_Data.apply(result,axis=1)
IPL_Final_Data['Team1Home'] = IPL_Final_Data.apply(T1HomeGround,axis=1)
IPL_Final_Data['Team2Home'] = IPL_Final_Data.apply(T2HomeGround,axis=1)
IPL_Final_Data['BattingFirst'] = IPL_Final_Data.apply(battingfirst,axis=1)
```

```python
IPL_Final_Data.drop(['HomeGround_x','HomeGround_y'],axis=1,inplace=True)
IPL_Final_Data
```

```python
file5 = pd.read_csv("D:\IPL_WinPrediction\IPL_HeadToHead.csv")
file5
```

```python
IPL_Final_Data = pd.merge(IPL_Final_Data, file5, left_on=['Team1', 'Team2'], right_on=['Team1', 'Team2'])
IPL_Final_Data
```

```python
def head_to_head(row):
    if row['Matches']!=0:
        row['Team1WinPercent'] = row['Team1Wins']/row['Matches']
        row['Team2WinPercent'] = row['Team2Wins']/row['Matches']
    else:
        row['Team1WinPercent'] = 0
        row['Team2WinPercent'] = 0
    return row
```

```python
IPL_Final_Data = IPL_Final_Data.apply(head_to_head, axis=1)
IPL_Final_Data
```
